# Problema Propuesto

En cierto país donde la enfermadad X es pandemica, se sabe que un 12% de la población padece de dicha enfermeras. Se dispone de una prueba para detectar la enfermedad, pero no es totalmente fiable, ya que , da positivo en el 90 % de los casos de personas realmente enfermas, y da positivo en el 5 % de personas sanas.

P(S,P) / P(P)
(5/100)(88/100) / ((5/100)(88/100) + (90/100)(12/100)= 0.2895

## Calculadora en python

```python
def bayes(p_a, p_b_given_a, p_b_given_not_a):
    # probabilidad de A
    not_a = 1 - p_a
    # probabilidad de B
    p_b = p_b_given_a * p_a + p_b_given_not_a * not_a
    # probabilidad de A dado B
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b
 
# P(A)
p_a = 0.12
# P(B|A)
p_b_given_a = 0.88
# P(B|A)
p_b_given_not_a = 0.2895
# calculo de P(A|B)
result = bayes(p_a, p_b_given_a, p_b_given_not_a)
# suma
print('P(A|B) = %.3f%%' % (result * 100))
```
